```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
loans = pd.read_csv('loan_data.csv')
```

```python
loans.head()
loans.info()
loans.describe()
```

```python
plt.figure(figsize=(10,6))

loans[loans['credit.policy']==1]['fico'].hist(
    bins=30, color='blue', alpha=0.5, label='Credit.Policy=1'
)

loans[loans['credit.policy']==0]['fico'].hist(
    bins=30, color='red', alpha=0.5, label='Credit.Policy=0'
)

plt.legend()
plt.xlabel('FICO')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10,6))

loans[loans['not.fully.paid']==1]['fico'].hist(
    bins=30, color='blue', alpha=0.5, label='not.fully.paid=1'
)

loans[loans['not.fully.paid']==0]['fico'].hist(
    bins=30, color='red', alpha=0.5, label='not.fully.paid=0'
)

plt.legend()
plt.xlabel('FICO')
plt.ylabel('Frequency')
```

```python
sns.countplot(x='purpose', data=loans, hue='not.fully.paid')
```

```python
sns.lmplot(x='fico', y='int.rate', data=loans, hue='credit.policy',
           col='not.fully.paid')
```

```python
cat_feats = ['purpose']
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_pred))
print(classification_report(y_test, rfc_pred))
```

What performed better the random forest or the decision tree?

The Random Forest performed better than the Decision Tree, because Random Forest combines multiple decision trees, which reduces overfitting and improves accuracy, while a single decision tree can easily overfit the data.
